```python
y1 = []
y2 = []
y3 = []
x1 = []
x2 = []
with open("eval_result.csv") as f:
    for line in f:
        a, b, c, d, e = [float(x) for x in line.split(',')]
        e = int(e)
        y1.append(a)
        y2.append(b)
        y3.append(c)
        x1.append(d)
        x2.append(e)
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
plt.plot(x1[1::16], y1[1::16], label="F")
plt.plot(x1[1::16], y2[1::16], label="Percision")
plt.plot(x1[1::16], y3[1::16], label="Recall")
plt.legend()
```
