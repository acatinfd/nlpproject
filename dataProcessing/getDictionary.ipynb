```python
import pickle
import re
all_business = pickle.load(open( "../../saveBusiness.p", "rb" ))
```

```python
businessCount = 0
reviewCount = 0
tipCount = 0
tipWordDictionary = {}
reviewWordDictionary = {}
wordDictionary = {}
rgx = re.compile("(\w[\w']*\w|\w)")

for business in all_business.values():
    businessCount = businessCount + 1
    for tip in business['tips']:
        tipCount += 1
        tipWords = rgx.findall(tip.lower())
        for tWord in tipWords:
            if not tWord in tipWordDictionary:
                tipWordDictionary[tWord] = 1
            else:
                tipWordDictionary[tWord] += 1
            wordDictionary[tWord] = tipWordDictionary[tWord]
    
    for key in business['reviews']:
        reviewCount += 1
        review = key['text']
        reviewWords = rgx.findall(review.lower())
        
        for rWord in reviewWords:
            if not rWord in reviewWordDictionary:
                reviewWordDictionary[rWord] = 1
            else:
                reviewWordDictionary[rWord] += 1
            if not rWord in wordDictionary:
                wordDictionary[rWord] = reviewWordDictionary[rWord]
            else:
                wordDictionary[rWord] += reviewWordDictionary[rWord]
            
        
print ("tip words count: ", len(tipWordDictionary))
print ("review words count: ", len(reviewWordDictionary))
print ("total words count: ", len(wordDictionary))


example = 0
for key in tipWordDictionary:
    example += 1
    print (key, tipWordDictionary[key])
    if example > 100:
        break
```

```python
pickle.dump( tipWordDictionary, open( "saveTipDictionary.p", "wb") )
pickle.dump( reviewWordDictionary, open( "saveReviewDictionary.p", "wb") )
pickle.dump( wordDictionary, open( "saveTotalWordDictionary.p", "wb") )
```
